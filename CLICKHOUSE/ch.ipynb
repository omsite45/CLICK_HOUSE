we will replace contents of client.py file with below ones:

import clickhouse_connect
from .ingest import IngestionMixin
class ClickHouseAI(IngestionMixin):

    def __init__(
        self,
        host="localhost",
        port=8123,
        database="default",
        user="default",
        password="",
    ):
        self.client = clickhouse_connect.get_client(
            host=host,
            port=port,
            username=user,
            password=password,
            database=database,
        )

    def query(self, sql: str):
        result = self.client.query(sql)

        return {
            "columns": result.column_names,
            "rows": result.result_rows,
            "row_count": len(result.result_rows),
        }
