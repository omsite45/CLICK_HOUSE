```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("Retail Dataset") \
    .getOrCreate()

reviews = spark.read.csv(
    "datasets/reviews.csv",
    header=True,
    inferSchema=True
)

reviews.createOrReplaceTempView("reviews")

spark.sql("SELECT * FROM reviews where verified_purchase ='false' and review_date > '2025-01-01'").show()
```
